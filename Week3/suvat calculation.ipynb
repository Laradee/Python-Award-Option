```python
def main():
    '''calucator for suvat'''
    menu = '''suvat calucator
    1. calculate displacement
    2. calculate intial velocity
    3. calculate final velocity
    4. calculate acceleration
    5. calculate time '''
while True:
    print(menu)
    choice = int(input("choose a menu:"))


if choice == 4:
    def acceleration(): 
        ini = float(input("enter intial velocity:"))
        final = float(input("enter final velocity:"))
        tim = float(input("enter time taken:"))
        acc = (ini+final)/tim
        print(f"accelration is {acc} m/s^2")

if choice == 1:
    def displacement():
        ini = float(input("enter intial velocity:"))
        final = float(input("enter final velocity:"))
        tim = float(input("enter time taken:"))
        dis = 0.5*(ini+final)*tim
        print(f"displacement is {dis} m")

if choice == 5:
    def time():
        ini = float(input("enter intial velocity:"))
        final = float(input("enter final velocity:"))
        acc = float(input("enter acceleration:"))
        tim = (final-ini)/acc
        print(f"time is {tim} s")

if choice == 3:
    def final():
        ini = float(input("enter intial velocity:"))
        acc = float(input("enter acceleration:"))
        tim = float(input("enter time taken:"))
        final = ini+(acc*tim)
        print(f"final velocity is {final} m/s")

if choice == 2:
    def ini():
        final = float(input("enter final velocity:"))
        acc = float(input("enter acceleration:"))
        tim = float(input("enter time taken:"))
        ini = final-(acc*tim)
        print(f"intial velocity is {ini} m/s")

```
